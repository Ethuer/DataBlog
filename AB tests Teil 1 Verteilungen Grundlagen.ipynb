```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import random 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plot 2 distributions
normal = pd.DataFrame()

x_rand = [np.random.normal(1,2) for x in range(10)]
y_rand = [np.random.normal(1,2) for x in range(10)]

normal['x'] = x_rand
normal['y'] = y_rand

sns.distplot(normal[['x']], hist=True, rug=True)
sns.distplot(normal[['y']], hist=False, rug=True).set_title('2 Idente verteilungen sehen sich nicht immer ähnlich')
```

```python
f, axes = plt.subplots(1, 2, figsize=(15,4))

# plot 2 distributions

normal = pd.DataFrame()

x1_rand = [np.random.normal(1,2) for x in range(6)]
y1_rand = [np.random.normal(1,2) for x in range(6)]

x2_rand = [np.random.exponential(2) for x in range(6)]
y2_rand = [np.random.exponential(3) for x in range(6)]


normal['x1'] = x1_rand
normal['x2'] = x2_rand
normal['y1'] = y1_rand
normal['y2'] = y2_rand

sns.distplot(normal[['x1']], hist=True, rug=True, ax=axes[0])
sns.distplot(normal[['x2']], hist=False, rug=True, ax=axes[0]).set_title('2 Idente verteilungen sehen sich nicht immer ähnlich')
sns.kdeplot(normal.x1, normal.x2, ax=axes[1], shade=True,shade_lowest=False)
sns.kdeplot(normal.x2, normal.y2, ax=axes[1], shade=True, shade_lowest=False).set_title('Gleiche Verteilung in 2D')
```

```python
# create some data from a normal distribution

normal = pd.DataFrame()

x_rand = [np.random.normal(1,2) for x in range(200)]
y_rand = [np.random.normal(1,2) for x in range(200)]

normal['x'] = x_rand
normal['y'] = y_rand
```

```python
# points between - ~2.3 and + ~2.3
f, axes = plt.subplots(1, 2, figsize=(15,4))
sns.distplot(normal.x, ax=axes[0]).set_title('1D Normal Verteilung')
sns.kdeplot(normal.x, normal.y, ax=axes[1], shade=True).set_title('2D Normal Verteilung')
```

```python
# create some data from a normal distribution

exp = pd.DataFrame()

# exponential needs to know beta (= 1/lambda ) and size 
x_rand = [np.random.exponential(5) for x in range(200)]
y_rand = [np.random.exponential(3) for x in range(200)]

exp['x'] = x_rand
exp['y'] = y_rand

# points between - ~2.3 and + ~2.3
f, axes = plt.subplots(1, 2, figsize=(15,4))
sns.distplot(exp.x, ax=axes[0]).set_title('1D Exponential Verteilung')
sns.kdeplot(exp.x, exp.y, ax=axes[1], shade=True).set_title('2D Exponential Verteilung')
```

```python
# create some data from a normal distribution

poisson = pd.DataFrame()

# poisson needs to know its lambda 
x_rand = [np.random.poisson(5) for x in range(200)]
y_rand = [np.random.poisson(3) for x in range(200)]

poisson['x'] = x_rand
poisson['y'] = y_rand

# points between - ~2.3 and + ~2.3
f, axes = plt.subplots(1, 2, figsize=(15,4))
sns.distplot(poisson.x, ax=axes[0]).set_title('1D Poisson Verteilung')
sns.kdeplot(poisson.x, poisson.y, ax=axes[1], shade=True).set_title('2D Poisson Verteilung')
```

```python
# create some data from a normal distribution

beta = pd.DataFrame()

# beta wants two parameters , alpha and beta
x_rand = [np.random.beta(5,2) for x in range(200)]
y_rand = [np.random.beta(3,2) for x in range(200)]

beta['x'] = x_rand
beta['y'] = y_rand

# points between - ~2.3 and + ~2.3
f, axes = plt.subplots(1, 2, figsize=(15,4))
sns.distplot(beta.x, ax=axes[0]).set_title('1D Beta Verteilung')
sns.kdeplot(beta.x, beta.y, ax=axes[1], shade=True).set_title('2D Beta Verteilung')
```
